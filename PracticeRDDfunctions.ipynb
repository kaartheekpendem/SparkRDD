```
rdd=sc.textFile("/FileStore/tables/weather.csv",2)


rdd_year=rdd.map(lambda i:(int(i.split(",")[0].split("-")[0]),int(i.split(",")[2])))
rdd_year_temp=rdd_year.reduceByKey(lambda x,y:x if x>y else y)
srt=rdd_year_temp.sortBy(lambda x:x[1])
print("Max tmp {}".format(rdd_year_temp.collect()))

print("format {}".format(rdd_year_temp.partitioner.partitionFunc))
```
