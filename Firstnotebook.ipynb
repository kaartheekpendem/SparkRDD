```
i=[2,1,3,8,4,5,6,7,10]
rdd1=sc.parallelize(i)
lamda_rdd1=rdd1.sortBy(lambda i:i).collect()
print(lamda_rdd1)
```

```
square_rdd=rdd1.map(lambda i:i**2).collect()
print(square_rdd)
```

```
reduce_rdd=rdd1.reduce(lambda x,y: x if x>y else y)
print(reduce_rdd)
```
