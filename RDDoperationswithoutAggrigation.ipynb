```
def startswith5(indicator):
  return indicator.startswith("s")
```

```
findwords=rdd.filter(lambda c: startswith5(c))
display(findwords.collect())
```

```
x= rdd.map(lambda c: (c,c[0],c.startswith("s")))
display(x.collect())
```

```
y=x.filter(lambda v: v[2])
display(y.take(5))
```

```
r=sc.parallelize(range(1,21,2))
b=r.reduce(lambda x,y: x+y)
print(b)
```

```
def wordbylen(lw,rw):
  if len(lw)>len(rw):
    return lw
  else:
    return rw
  
rdd.reduce(wordbylen)
```

```
def partword(iter):
  result=[]
  result.append("Starting")
  result+=[(len(i),i) for i in iter]
  result.append("closed")
  return result


g=rdd.mapPartitions(partword)
for i in g.collect():
  print(i)
```

```
def indexofrdd(s,j):
  return ["partition {}=>{}" .format(s,x) for x in j]

rdd.mapPartitionsWithIndex(indexofrdd).collect()
```
